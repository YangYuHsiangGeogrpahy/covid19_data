```python
import pandas as pd
def download_covid(select,pth):
    if select not in ['combined','latest','responses','vaccines']:
        print ('please select from combined,latgest,responses,vaccines')
    if select=='combined':
        url = "https://github.com/OxCGRT/covid-policy-tracker/blob/master/data/OxCGRT_latest_combined.csv?raw=true"
    elif select=='latest':
        url = "https://github.com/OxCGRT/covid-policy-tracker/blob/master/data/OxCGRT_latest.csv?raw=true"
    elif select=='responses':
        url = "https://github.com/OxCGRT/covid-policy-tracker/raw/master/data/OxCGRT_latest_responses.csv?raw=true"
    elif select=='vaccines':
        url = "https://github.com/OxCGRT/covid-policy-tracker/raw/master/data/OxCGRT_vaccines_full.csv?raw=true"
    df = pd.read_csv(url)
    # Date 轉為日期格式 date
    if 'Date' in list(df.columns):
        df['date'] = pd.to_datetime(df['Date'], format="%Y%m%d")
    print (df.shape)
    print (df.columns)
    fname=pth +'OxCGRT_'+select+'.csv'
    df.to_csv( fname)
    return df
```

## single_country_multi_variables(df,start, end, field, country)
* 一個國家多個指標一個國家多個指標
* df 為資料 dataframe
* start 為起始日
* end 為結束日
* field 為變數欄位 list
* country 為list of string 格式

```python
# 一個國家多個指標
def single_country_multi_variables(df,start, end, field, country):
    if type(country)==str:
        mask1=df.CountryName == country
    elif type(country)==list:
        mask1=df.CountryName == country[0]   # 因為為list格式
    mask2=(df.Date>=start) & (df.Date<=end)
    df1=df[['date']+field][mask1][mask2]
    # 以 date 為 index
    df1.index=list(df1.date)
    del df1['date']
    return df1
```

## 數值為 nan 其值設定為與前一天同
* 檔案大，需要一點時間

```python
import numpy as np
country=df.CountryName.unique()
column=df.columns
df1=pd.DataFrame()
for c in range(len(df.CountryName.unique())):
    df_country=df[df.CountryName==country[c]]
    # 逐筆檢查
    for n in range(len(df_country)):
        # 有 nan 的欄位
        na =df_country.iloc[n][df_country.iloc[n].isna()].index
        if len(na)>0:
            for m in range(len(na)):
                # 第一筆若為 nan 設為 0
                if n==0:
                    df_country[na[m]].iloc[n]=0
                else:  # 否則與前一筆一樣
                    df_country[na[m]].iloc[n]=df_country[na[m]].iloc[n-1]
    df1=df1.append(df_country)
df1
```

```python
import numpy as np
# 各國分開處理，第一筆 nan 因為無增加率
previous=list(df["cases"])
previous.insert(0, 1)
previous.pop(-1)
# 前一天的資料
df['previous']=previous
# 空白的 dataframe
df1=pd.DataFrame()
country=df.CountryName.unique()
# 各國分別計算
for c in range(len(df.CountryName.unique())):
    df_country=df[df.CountryName==country[c]]
    # 計算增加率，目前的減前一天的，如果資料有誤，相減為負呢？
    df_country["case_inc"]=(df["cases"]-df["previous"])
    # 第一筆為 0
    df_country["case_inc"].iloc[0]=0
    # 整併至 df1
    df1=df1.append(df_country)
df=df1.copy()
df
```

## plt_single_variable_multi_country_trend()
* 多個國家，多個變數，折線圖，
* df 為 dataframe，每日各國各個變數

```python
import matplotlib.pyplot as plt
def plt_single_variable_multi_country_trend(df,date, country,variable,rolling):
    plt.figure(figsize=(15,8))
    # 單一變數，多個國家
    if type(country)==str:
        country=[country]
    if type(variable)==list:
        variable=variable[0]
    for c in range(len(country)):
        mask = df["CountryName"]== country[c]
        df1=df[mask]
        x=df1[date].unique()    # date 為日期格式
        plt.xticks(np.arange(0, len(x), step=80),rotation=50)
        # 繪圖
        plt.plot(x,df1[variable].rolling(rolling).mean(),label=country[c])
    plt.legend(loc=0)
    plt.title(variable)
    plt.show()
```

## plt_direct_vari_trend(df,field,rolling)
* df 單一變數，欄位為國家，列為日期
* field 欄位名稱
* rolling 前面n個平均

```python
import matplotlib.pyplot as plt
def plt_direct_vari_trend(df,field,rolling):
    if type(field)==str:
        field=[field]
    plt.figure(figsize=(15,8))
    x=df.index
    plt.xticks(list(np.arange(0, len(x), step=len(x)/10)),rotation=50)
    for v in range(len(field)):
        plt.plot(x,df[field[v]].rolling(rolling).mean(),label=field[v])
    plt.legend(loc=0)
    plt.title(field)
    plt.show()
```

## plot_cate() 依不同類別，顯示變數差異
* 單一變數
* 多個分類

```python
def plot_cate(cat_var,var,df):
    category=df[cat_var].unique()
    final=[]
    for cat in category:
        dfc=df[df[cat_var]==cat]
        final.append(dfc[var].mean())
    dfp=pd.DataFrame(final, columns=[var], index=category)
    dfp.plot(kind="bar")
    return dfp
```

```python
var='case_inc_p'
dfp=plot_cate("region2",var,df)
```

```python
var='death_inc_p'
dfp=plot_cate("region2",var,df)
```

```python
var='demo'
dfp=plot_cate("region2",var,df)
```
